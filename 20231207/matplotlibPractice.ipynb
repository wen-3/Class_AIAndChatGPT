* 專題 - 圖書分類銷售分析圖

```python
dataM = [14, 16, 8, 13, 16, 12, 16, 5]
dataF = [10, 19, 6, 10, 13, 13, 20, 9]
labels = ['商業理財', '文學小說', '藝術設計', '人文科普', '語言電腦', '心靈養生', '生活風格', '親子共享']

plt.figure(figsize=[10, 4])
plt.subplot(1, 2, 1)
plt.title('圖表分類銷售比率-男性')
plt.pie(dataM,
        labels=labels,
        labeldistance=1.1,
        autopct="%2.1f%%",
        startangle=0)

plt.subplot(1, 2, 2)
plt.title('圖表分類銷售比率-女性')
plt.pie(dataF,
        labels=labels,
        labeldistance=1.1,
        autopct="%2.1f%%",
        startangle=0)

plt.rcParams["font.sans-serif"] = "Microsoft JhengHei"
plt.rcParams["axes.unicode_minus"] = False

plt.show()
```

```python
import matplotlib.pyplot as plt
x = ['商業理財', '文學小說', '藝術設計', '人文科普', '語言電腦', '心靈養生', '生活風格', '親子共享']
dataM = [14, 16, 8, 13, 16, 12, 16, 5]
dataF = [10, 19, 6, 10, 13, 13, 20, 9]

width=0.3
listx_M = [x - width/2 for x in range(len(x))]
listx_F = [x + width/2 for x in range(len(x))]

plt.bar(listx_M, dataM, width)
plt.bar(listx_F, dataF, width)
plt.legend(['男', '女'], loc='upper left')
# plt.legend(['男', '女'], loc=2)

plt.title('圖書分類銷售長條圖-性別')
plt.xlabel('圖書分類')
plt.ylabel('銷售比率(%)')
plt.xticks(range(len(x)), x)

plt.rcParams["font.sans-serif"] = "Microsoft JhengHei"
plt.rcParams["axes.unicode_minus"] = False

plt.show()
```
