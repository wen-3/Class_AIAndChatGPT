```python
se1 = pd.Series({'王小明':45, '李小美':32, '陳大同':78, '林小玉':56})
se2 = pd.Series({'王小明':67, '李小美':98, '陳大同':67, '林小玉':37})
se3 = pd.Series({'王小明':87, '李小美':67, '陳大同':56, '林小玉':38})
se4 = pd.Series({'王小明':67, '李小美':56, '陳大同':45, '林小玉':39})
se5 = pd.Series({'王小明':28, '李小美':34, '陳大同':34, '林小玉':35})

df = pd.DataFrame({'國文':se1, '英文':se2, '數學':se3, '自然':se4, '社會':se5})
df
```

* 使用 concat 函數合併 Series 物件
    * axis = 0 => 直向合併

```python
se1 = pd.Series({'王小明':45, '李小美':32, '陳大同':78, '林小玉':56})
se2 = pd.Series({'王小明':67, '李小美':98, '陳大同':67, '林小玉':37})
se3 = pd.Series({'王小明':87, '李小美':67, '陳大同':56, '林小玉':38})
se4 = pd.Series({'王小明':67, '李小美':56, '陳大同':45, '林小玉':39})
se5 = pd.Series({'王小明':28, '李小美':34, '陳大同':34, '林小玉':35})

df = pd.concat([se1, se2, se3, se4, se5], axis=1)
df.columns = ['國文', '英文', '數學', '自然', '社會']
df
```
